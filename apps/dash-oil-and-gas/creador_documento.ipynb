```python
import pandas as pd
from docx import Document
from docx.shared import Inches
```

```python
df = pd.DataFrame({
    'Name': ['John', 'Alice', 'Bob', 'Jane', 'Eva', 'Mark', 'Mike', 'Lucy'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'Maths': [80, 90, 70, 85, 75, 90, 60, 95],
    'Science': [75, 80, 65, 90, 70, 80, 70, 85],
    'History': [85, 75, 80, 90, 60, 70, 75, 80]
})
```

```python
df_long = pd.melt(df, id_vars=['Name', 'Gender'], var_name='Subject', value_name='Score')

# Pivot the data into a wide format
df_wide = pd.pivot_table(df_long, values='Score', index=['Name', 'Gender'], columns='Subject').reset_index()
```

```python
document = Document()

# Add a heading to the document
document.add_heading('Average Score by Subject', level=0)

# Add a table to the document
table = document.add_table(rows=1, cols=4)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Subject'
hdr_cells[1].text = 'Average Score'
hdr_cells[2].text = 'Male Average'
hdr_cells[3].text = 'Female Average'

# Group the data by Subject and calculate the mean scores
grouped = df_long.groupby('Subject').mean()

for index, row in grouped.iterrows():
    # Add a row to the table
    row_cells = table.add_row().cells
    row_cells[0].text = index
    row_cells[1].text = str(row['Score'])

    # Calculate the mean scores by gender
    male_avg = df_wide[df_wide['Gender'] == 'Male'][index].mean()
    female_avg = df_wide[df_wide['Gender'] == 'Female'][index].mean()
    
    row_cells[2].text = str(round(male_avg, 2))
    row_cells[3].text = str(round(female_avg, 2))

# Add a paragraph to the document
document.add_paragraph('The average score by subject is shown in the table above.')

# Save the document
document.save('report.docx')
```

```python
from docx import Document
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import RGBColor, Pt

document = Document()

# Add a table to the document
table = document.add_table(rows=1, cols=4)
table.alignment = WD_TABLE_ALIGNMENT.CENTER

# Add table header and set style
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Subject'
hdr_cells[1].text = 'Average Score'
hdr_cells[2].text = 'Male Average'
hdr_cells[3].text = 'Female Average'
for cell in hdr_cells:
    cell.paragraphs[0].runs[0].bold = True
    cell.paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
    cell.width = 1.5

# Set the background color of the header cells
for cell in hdr_cells:
    shading_elm = cell._element.xpath('.//w:shd')[0]
    shading_elm.set('fill', '0066CC')
    shading_elm.set('themeFill', 'text1')
    shading_elm.set('themeFillTint', '50')

# Group the data by Subject and calculate the mean scores
grouped = df_long.groupby('Subject').mean()

# Add rows to the table and set style
for index, row in grouped.iterrows():
    row_cells = table.add_row().cells
    row_cells[0].text = index
    row_cells[1].text = str(round(row['Score'], 2))
    male_avg = df_wide[df_wide['Gender'] == 'Male'][index].mean()
    female_avg = df_wide[df_wide['Gender'] == 'Female'][index].mean()
    row_cells[2].text = str(round(male_avg, 2))
    row_cells[3].text = str(round(female_avg, 2))
    for cell in row_cells:
        cell.width = 1.5
        cell.paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER

# Set the font color of the cells in the first column to blue
for i in range(1, len(table.rows)):
    cell = table.cell(i, 0)
    cell.paragraphs[0].runs[0].font.color.rgb = RGBColor(0, 102, 204)

# Add a border to the cells in the table
for row in table.rows:
    for cell in row.cells:
        for p in cell.paragraphs:
            for r in p.runs:
                r.font.size = Pt(10)
                r.font.name = 'Calibri'
        tc = cell._element.tcPr
        tc.append(cell._element.xpath('.//w:tcBorders')[0])

# Add a paragraph to the document
document.add_paragraph('The average score by subject is shown in the table above.')

# Save the document
document.save(r'C:\Users\Ministerio\Desktop\Decreto-277\report.docx')
```
